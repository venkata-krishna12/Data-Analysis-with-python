```python
import matplotlib.pyplot as plt
plt.scatter(df["age"], df["smoker"].apply(lambda x: 1 if x == "yes" else 0), c="blue", label="Smoker (1=Yes, 0=No)")
plt.title("Scatter Plot: Age vs Smoking")
plt.xlabel("Age")
plt.ylabel("Smoker Status (1=Yes, 0=No)")
plt.axhline(0.5, color="gray", linestyle="--", alpha=0.7)
plt.legend()
plt.show()
```

```python
#Filter data and plot (Age between 20 and 40)
filtered_df = df[(df["age"] >= 20) & (df["age"] <= 40)]
plt.scatter(filtered_df["age"], filtered_df["smoker"].apply(lambda x: 1 if x == "yes" else 0), c="green", label="Filtered Data")
plt.title("Scatter Plot: Age (20-40) vs Smoking")
plt.xlabel("Age")
plt.ylabel("Smoker Status (1=Yes, 0=No)")
plt.axhline(0.5, color="gray", linestyle="--", alpha=0.7)
plt.legend()
plt.show()
```

```python
#Expenses of smokers vs non-smokers
expenses_smokers = df[df["smoker"] == "yes"]["expenses"]
expenses_non_smokers = df[df["smoker"] == "no"]["expenses"]
plt.bar(["Smokers", "Non-Smokers"], [expenses_smokers.mean(), expenses_non_smokers.mean()], color=["red", "blue"])
plt.title("Average Expenses: Smokers vs Non-Smokers")
plt.ylabel("Expenses")
plt.show()
```

```python
#Number of smokers gender-wise
smokers_gender = df[df["smoker"] == "yes"]["sex"].value_counts()
smokers_gender.plot(kind="bar", color=["purple", "orange"])
plt.title("Number of Smokers by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Scatter plot between age and expenses
plt.scatter(df["age"], df["expenses"], c="orange", alpha=0.7)
plt.title("Scatter Plot: Age vs Expenses")
plt.xlabel("Age")
plt.ylabel("Expenses")
plt.show()
```
